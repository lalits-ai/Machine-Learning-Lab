```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression

import warnings
def warn(*args, **kwargs):
    pass
warnings.warn = warn
```

```python
df = pd.read_csv(r"C:\Jupyter Projects\Machine Learning\ML Lab\Week 3\Exercises\studyHours.csv")
df
```

```python
plt.scatter(df['Hours'], df['Scores'])
```

```python
X = df.Hours
y = df.Scores
```

```python
X = np.array(X).reshape(-1, 1)
X.shape
```

```python
y.shape
```

```python
linearReg = LinearRegression()
```

```python
linearReg.fit(X, y)
```

```python
coeff = linearReg.coef_
intercept = linearReg.intercept_
```

```python
coeff
```

```python
intercept
```

```python
df.plot(kind='scatter', x='Hours', y='Scores')
plt.plot(X, linearReg.predict(X), color='red')
```

```python
from sklearn.metrics import r2_score
print(f"R2 score: {r2_score(y, linearReg.predict(X))}")
```
