```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
l1 = [2,9,5,5,3,7,1,8,8,2]
l2 = [69,98,82,77,71,84,55,94,84,64]

data = {'Hours_studied' : l1,
        'grades' : l2}

df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(14, 10))
plt.subplot(2,3,1)
plt.hist(df['Hours_studied'], bins=5, color='blue', alpha=.5)

plt.subplot(2,3,2)
plt.bar(df['Hours_studied'], df['grades'], color='red')

plt.subplot(2,3,3)
pie_label = ['Grade' + str(i+1) for i in range(len(df['grades']))]
plt.pie(df['grades'], labels=pie_label, autopct='%1.1f%%', startangle=90)

plt.subplot(2,3,4)
plt.scatter(df['Hours_studied'], df['grades'], color='green')

plt.subplot(2,3,5)
plt.boxplot([df['Hours_studied'], df['grades']])

plt.show()
```
