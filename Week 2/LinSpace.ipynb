# Explore linspace function

numpy.linspace() is a powerful function in Python’s NumPy library that allows you to create evenly spaced arrays with a specified number of elements. It’s a versatile tool used in various applications, from scientific computing to data analysis and data visualization.

```python
import numpy as np
```

```python
'''
Syntax 1 - np.linspace(start, stop)
Syntax 2 - np.linspace(start, stop, num, endpoint, dtype, axis)

num - number of samples to be generated
endpoint - (default=True) Inclusion of the stop value
axis - (default=0) Axis to insert the generated samples
'''

# Generate 50 samples between the interval 1 and 20
data = np.linspace(1, 20, num=50)
data
```

```python
# Generate 5 samples between the interval 2 and 100 and also include the stop value
data = np.linspace(2, 100, num=5, endpoint=True, dtype='float64')
data
```
