### Using Matplotlib, perform the following:

* Simple Scatter Plot
* Sine curve plot
* Data generation using random function
* Histograms, bar plots, box plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import random
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
print(x)
print(y)
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.scatter(x, y)
ax.set(title='Spread of x and y', xlabel='x', ylabel='y')
plt.show()
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.plot(x, y)
ax.set(title='Spread of x and y', xlabel='x', ylabel='y')
plt.show()
plt.tight_layout()
```

```python
import math
new_x = np.arange(0, math.pi*2, 0.5)
new_y = np.sin(new_x)
```

```python
plt.scatter(new_x, new_y, color='green')
plt.legend(['datapoints'])
for xy in zip(new_x, new_y):
    plt.annotate('(%.2f, %.2f)'%xy, xy=xy)
```

```python
plt.bar(new_x, new_y, color='red')
```

```python
class_his = {'C++':33,'DSA':45,'History':98}
courses = list(class_his.keys())
numbers = list(class_his.values())
```

```python
plt.bar(courses, numbers, width=.4)
```

```python
x = random.sample(range(10,20), 5)
y = random.sample(range(20,30), 5)
plt.subplot(1,2,1)
plt.plot(x,y)

x = random.sample(range(10,20), 5)
y = random.sample(range(20,30), 5)
plt.subplot(1,2,2)
plt.plot(x, y)

plt.show()
```

```python
data = np.random.rand(100)
new_data = np.arange(1, 10, .2)
data
new_data

updated_data = np.append(data, new_data, axis=0)
updated_data
```

```python
plt.boxplot(data)
plt.title("Boxplot of data")
plt.xlabel("data")
plt.ylabel("values")
plt.show()
```

```python
plt.boxplot(updated_data)
plt.title("Newly appended data with outliers")
plt.show()
```
